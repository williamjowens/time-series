# Preprocess the Data

```python
def create_percent_change(df, target):
    """
    Create a 'percent_change' column for the target variable.

    Parameters:
    - df: Pandas DataFrame.
    - target: String, name of the target column.

    Returns:
    - DataFrame with an added 'percent_change' column.
    """
    df[f'{target}_percent_change'] = df[target].pct_change() * 100
    # Replace infinite values with 0
    df[f'{target}_percent_change'].replace([np.inf, -np.inf], 0, inplace=True)
    return df

# Apply 'percent_change' to 'strikes'
strikes_data = create_percent_change(strikes_data, 'strikes')
```

```python
def create_spike_indicator(df, target, quantile=0.95):
    """
    Create a binary 'spike' indicator where values above a certain quantile are marked as spikes.

    Parameters:
    - df: Pandas DataFrame.
    - target: String, name of the target column.
    - quantile: Float, quantile to define spike threshold.

    Returns:
    - DataFrame with an added 'spike' column.
    """
    spike_threshold = df[target].quantile(quantile)
    df['spike'] = (df[target] > spike_threshold).astype(int)
    return df

# Apply 'spike' indicator to 'strikes'
strikes_data = create_spike_indicator(strikes_data, 'strikes')
```

```python
def create_zero_count_indicator(df, target):
    """
    Create a binary 'zero_count' indicator where values are marked as 0.

    Parameters:
    - df: Pandas DataFrame.
    - target: String, name of the target column.

    Returns:
    - DataFrame with an added 'zero_count' column.
    """
    df['zero_count'] = (df[target] == 0).astype(int)
    return df

# Apply 'zero_count' indicator to 'strikes'
strikes_data = create_zero_count_indicator(strikes_data, 'strikes')
```

```python
# Verify derived variables
print("After Creating Derived Variables:")
display(strikes_data[['strikes', 'strikes_percent_change', 'spike', 'zero_count']].head(5))
```

```python
def create_lagged_features(df, target, lags):
    """
    Create lagged features for a specified target variable.

    Parameters:
    - df: Pandas DataFrame.
    - target: String, name of the target column.
    - lags: Integer, number of lags to create.

    Returns:
    - DataFrame with added lagged features.
    """
    for lag in range(1, lags + 1):
        df[f'{target}_lag_{lag}'] = df[target].shift(lag)
    return df

# Apply lagging to 'strikes'
N = 2  # Number of lags
strikes_data = create_lagged_features(strikes_data, 'strikes', N)
```

```python
# Apply lagging to exogenous variables
# Define exogenous variables to lag
exog_vars_to_lag = ['output', 'strikes_percent_change', 'spike', 'zero_count']

# Check if all exogenous variables exist
missing_vars = [var for var in exog_vars_to_lag if var not in strikes_data.columns]
if missing_vars:
    raise ValueError(f"The following exogenous variables are missing from the dataset and cannot be lagged: {missing_vars}")

# Apply lagging to each exogenous variable
for var in exog_vars_to_lag:
    strikes_data = create_lagged_features(strikes_data, var, N)
```

```python
# Verify lagged variables
lagged_cols = [f'strikes_lag_{lag}' for lag in range(1, N+1)]
for var in exog_vars_to_lag:
    lagged_cols += [f'{var}_lag_{lag}' for lag in range(1, N+1)]
print("After Creating Lagged Variables:")
display(strikes_data[lagged_cols].head(5))
```

```python
def generate_fourier_terms(df, period, K):
    """
    Generate Fourier terms for capturing seasonality.

    Parameters:
    - df: Pandas DataFrame with a 'date' column.
    - period: Integer, number of periods in a complete seasonal cycle (e.g., 12 for monthly data).
    - K: Integer, number of Fourier harmonics.

    Returns:
    - DataFrame containing sine and cosine Fourier terms.
    """
    t = np.arange(len(df))
    fourier_terms = {}
    for k in range(1, K + 1):
        fourier_terms[f'sin_{k}'] = np.sin(2 * np.pi * k * t / period)
        fourier_terms[f'cos_{k}'] = np.cos(2 * np.pi * k * t / period)
    return pd.DataFrame(fourier_terms, index=df.index)
```

```python
# Define seasonality parameters
period = 12  # Monthly data with yearly seasonality
K = 1        # Number of Fourier harmonics

# Generate and concatenate Fourier terms
fourier_terms_df = generate_fourier_terms(strikes_data, period, K)
strikes_data = pd.concat([strikes_data, fourier_terms_df], axis=1)

# Verify Fourier Terms
print("After Generating Fourier Terms:")
display(strikes_data[[f'sin_{k}' for k in range(1, K+1)] + [f'cos_{k}' for k in range(1, K+1)]].head(5))
```

```python
def validate_data(df, dataset_name):
    """
    Validate that the DataFrame contains no NaN or infinite values.

    Parameters:
    - df: Pandas DataFrame.
    - dataset_name: String, name of the dataset (e.g., 'Training Data').

    Returns:
    - None. Prints validation results.
    """
    nan_counts = df.isna().sum()
    if nan_counts.any():
        print(f"Warning: {dataset_name} contains NaN values.")
        print(nan_counts[nan_counts > 0])
    else:
        print(f"{dataset_name} is clean.")

    # Check for infinite values in numeric columns
    numeric_df = df.select_dtypes(include=[np.number])
    if np.isinf(numeric_df.values).any():
        print(f"Warning: {dataset_name} contains infinite values.")
    else:
        print(f"{dataset_name} has no infinite values.")
```

```python
# Top-level function for cached binomial PMF
@lru_cache(maxsize=None)
def binom_logpmf_cached(j, y_prev, rho_i):
    """
    Cached function to compute the logarithm of the Binomial probability mass function.

    Parameters:
    - j (int): Number of successes.
    - y_prev (int): Number of trials.
    - rho_i (float): Probability of success.

    Returns:
    - float: Logarithm of the Binomial PMF value.
    """
    return binom.logpmf(j, y_prev, rho_i)

# Top-level function to compute total log probability for Poisson INAR using log-space
def compute_total_log_prob_pois_logspace(y_t, y_lags, rho, mu_t, max_j):
    """
    Compute the log probability for a single observation in Poisson INAR using log-space calculations.

    This function calculates the log probability of observing count `y_t` given the past counts `y_lags`,
    thinning parameters `rho`, and the Poisson innovation term `mu_t`. It considers all valid combinations
    of retained counts from each lag, up to a maximum threshold `max_j`.

    Parameters:
    - y_t (int): Observed count at time t.
    - y_lags (tuple of ints): Array of lagged counts.
    - rho (tuple of floats): Array of thinning parameters.
    - mu_t (float): Mean of the Poisson innovation term at time t.
    - max_j (int): Maximum number of retained counts to consider for each lag.

    Returns:
    - float: Log probability for observation t.
    """
    log_probs = []

    # Define the range of retained counts for each lag based on y_prev and max_j
    j_ranges = [range(0, min(int(y_prev), max_j) + 1) for y_prev in y_lags]

    # Iterate over all possible combinations of retained counts across lags
    for j_tuple in product(*j_ranges):
        j_total = sum(j_tuple)

        # Skip combinations where total retained counts exceed observed count
        if j_total > y_t:
            continue  # Invalid combination, skip

        log_p_combination = 0.0
        valid = True

        # Calculate the log probability for each retained count in the combination
        for j_i, y_prev, rho_i in zip(j_tuple, y_lags, rho):
            if j_i > y_prev:
                valid = False  # Invalid retained count, skip this combination
                break
            log_p_j_i = binom_logpmf_cached(j_i, y_prev, rho_i)
            if not np.isfinite(log_p_j_i):
                valid = False  # Non-finite log probability, skip
                break
            log_p_combination += log_p_j_i

        if not valid:
            continue  # Skip invalid combinations

        # Calculate the log probability of the Poisson innovation term
        log_p_k = poisson.logpmf(y_t - j_total, mu_t)
        if not np.isfinite(log_p_k):
            continue  # Invalid Poisson PMF, skip

        # Append the combined log probability
        log_probs.append(log_p_combination + log_p_k)

    if not log_probs:
        return np.log(1e-10)  # Prevent log(0) by assigning a very small probability

    # Use log-sum-exp for numerical stability
    return logsumexp(log_probs)
```

```python
class PoissonINARResults(GenericLikelihoodModelResults):
    """
    Custom results class for the Poisson INAR model, extending GenericLikelihoodModelResults.

    This class provides a Pandas Series for parameter estimates and a customized summary method.
    """
    def __init__(self, model, mlefit, param_names):
        """
        Initialize the PoissonINARResults object.

        Parameters:
        - model (GenericLikelihoodModel): The fitted Poisson INAR model.
        - mlefit: The Maximum Likelihood Estimation result.
        - param_names (list of str): Names of the parameters.
        """
        super().__init__(model, mlefit)
        self._param_names = param_names
        self.mlefit = mlefit

    @property
    def params(self):
        """
        Retrieve the parameter estimates as a Pandas Series with named indices.

        Returns:
        - pd.Series: Parameter estimates with names.
        """
        return pd.Series(self.mlefit.params, index=self._param_names)

    def summary(self, yname=None, xname=None, title=None, alpha=.05):
        """
        Generate a summary of the model results.

        Parameters:
        - yname (str, optional): Name of the dependent variable.
        - xname (list of str, optional): Names of the exogenous variables.
        - title (str, optional): Title of the summary.
        - alpha (float, optional): Significance level for confidence intervals.

        Returns:
        - Summary: A summary object containing parameter estimates and statistics.
        """
        if xname is None:
            xname = self._param_names
        return super().summary(yname=yname, xname=xname, title=title, alpha=alpha)
```

```python
class PoissonINAR(GenericLikelihoodModel):
    """
    Poisson INAR (Integer-valued Autoregressive) Model.

    This model is suitable for count data with autocorrelation, modeling the current count as a
    combination of thinned past counts and a Poisson-distributed innovation term.

    Attributes:
    - p (int): Number of autoregressive lags.
    - n_jobs (int): Number of parallel jobs for computation.
    - max_j (int): Maximum number of retained counts to consider per lag.
    - var_names (list of str): Names of the exogenous variables.
    """

    def __init__(self, endog, exog, p=1, n_jobs=-1, max_j=12, **kwds):
        """
        Initialize the Poisson INAR model.

        Parameters:
        - endog (array-like): The dependent variable (count data).
        - exog (array-like): The exogenous variables (predictors), excluding lagged counts.
        - p (int, optional): Number of autoregressive lags. Default is 1.
        - n_jobs (int, optional): Number of parallel jobs for computation. Default is -1 (all cores).
        - max_j (int, optional): Maximum number of retained counts to consider per lag. Default is 12.
        - **kwds: Additional keyword arguments for GenericLikelihoodModel.
        """
        self.p = p  # Number of AR terms
        self.n_jobs = n_jobs  # Number of parallel jobs
        self.max_j = max_j  # Maximum count to consider per lag

        # Extract variable names if exog is a DataFrame; else, generate generic names
        if isinstance(exog, pd.DataFrame):
            self.var_names = exog.columns.tolist()
        else:
            self.var_names = [f'x{i+1}' for i in range(exog.shape[1])]

        super(PoissonINAR, self).__init__(endog, exog, **kwds)

    def nloglikeobs(self, params):
        """
        Compute the negative log-likelihood for each observation.

        This function calculates the negative log-likelihood by summing the log probabilities
        of observing each count, given the past counts and exogenous variables.

        Parameters:
        - params (array-like): Parameter vector [beta coefficients..., gamma1, gamma2, ..., gamma_p].

        Returns:
        - np.ndarray: Negative log-likelihood values for each observation.
        """
        num_beta = self.exog.shape[1]
        beta = params[:num_beta]   # Regression coefficients
        gamma = params[num_beta:]  # Thinning parameters

        # Logistic transformation to ensure rho is between 0 and 1
        rho = 1.0 / (1.0 + np.exp(-gamma))  # rho = [rho1, rho2, ..., rho_p]

        # Linear predictor for Poisson mean
        mu = np.exp(np.dot(self.exog, beta))  # mu = exp(X * beta)

        # Prepare arguments for parallel processing
        obs_args = [
            (self.endog[t], self.endog[t - self.p:t], tuple(rho), mu[t])
            for t in range(self.p, len(self.endog))
        ]

        # Compute log probabilities in parallel using the 'threading' backend
        log_probs = Parallel(n_jobs=self.n_jobs, backend='threading')(
            delayed(compute_total_log_prob_pois_logspace)(y_t, y_lags, rho, mu_t, self.max_j)
            for y_t, y_lags, rho, mu_t in obs_args
        )

        # Check for any NaN or infinite values in log_probs
        if np.any(~np.isfinite(log_probs)):
            return np.full_like(log_probs, np.inf)  # Assign infinity to invalid computations

        return -np.array(log_probs)  # Return negative log-likelihood

    def fit(self, start_params=None, maxiter=1000, maxfun=5000, method='lbfgs', verbose=1, **kwds):
        """
        Fit the Poisson INAR model using Maximum Likelihood Estimation.

        Parameters:
        - start_params (array-like, optional): Initial guess for the parameters.
        - maxiter (int, optional): Maximum number of iterations for the optimizer. Default is 1000.
        - maxfun (int, optional): Maximum number of function evaluations. Default is 5000.
        - method (str, optional): Optimization method to use. Default is 'lbfgs'.
        - verbose (int, optional): Verbosity level. Default is 1 (display progress).
        - **kwds: Additional keyword arguments for the optimizer.

        Returns:
        - PoissonINARResults: Fitted model results with named parameters.
        """
        num_beta = self.exog.shape[1]
        num_rho = self.p

        # Initialize start_params if not provided
        if start_params is None:
            start_beta = np.zeros(num_beta)  # Starting values for beta coefficients
            start_gamma = np.zeros(num_rho)  # Starting values for gamma parameters
            start_params = np.concatenate([start_beta, start_gamma])

        # Define bounds for the parameters (no bounds in this case; adjust as needed)
        bounds = [(None, None)] * len(start_params)  # Example: no bounds

        # Fit the model using the superclass's fit method
        mlefit = super(PoissonINAR, self).fit(
            start_params=start_params,
            maxiter=maxiter,
            maxfun=maxfun,
            method=method,  # Optimizer method (e.g., 'lbfgs')
            bounds=bounds,
            disp=verbose > 0,  # Display optimizer output if verbose > 0
            **kwds
        )

        # Create parameter names: exogenous variables + gamma parameters
        param_names = self.var_names + [f'gamma_{i+1}' for i in range(num_rho)]

        # Return the custom results object with named parameters
        return PoissonINARResults(self, mlefit, param_names)

    def predict_custom(self, params, exog=None, last_p_y=None, random_state=None):
        """
        Generate predictions using the fitted Poisson INAR model.

        This method performs stochastic prediction by simulating the thinning process and Poisson innovations.

        Parameters:
        - params (array-like): Fitted parameter vector [beta coefficients..., gamma1, gamma2, ..., gamma_p].
        - exog (array-like): Exogenous variables for prediction (must align with training exog).
        - last_p_y (array-like): List or array of the last p observed counts from the training data to initialize predictions.
        - random_state (int, optional): Seed for the random number generator to ensure reproducibility. Default is None.

        Returns:
        - np.ndarray: Array of predicted y values.
        """
        num_beta = self.exog.shape[1]
        beta = params[:num_beta]  # Extract beta coefficients
        gamma = params[num_beta:]  # Extract gamma parameters

        # Logistic transformation to obtain thinning probabilities
        rho = 1.0 / (1.0 + np.exp(-gamma))  # rho = [rho1, rho2, ..., rho_p]

        # Calculate Poisson means for the innovation term
        mu = np.exp(np.dot(exog, beta))  # mu = exp(X * beta)

        # Ensure that last_p_y is provided for initializing the lags
        if last_p_y is None:
            raise ValueError("last_p_y (last p y's from training data) must be provided for prediction.")

        last_p_y = list(last_p_y)  # Make a mutable copy
        y_pred = []  # List to store predicted counts

        # Initialize the random number generator if a seed is provided
        if random_state is not None:
            rng = np.random.default_rng(random_state)
        else:
            rng = np.random.default_rng()

        # Iterate over each exogenous observation to generate predictions
        for i in range(len(exog)):
            y_lags = last_p_y.copy()  # Current lagged counts
            mu_t = mu[i]  # Poisson mean for the innovation term

            # Perform stochastic thinning for each lag
            carried_over = 0
            for j in range(self.p):
                carried_over += rng.binomial(y_lags[j], rho[j])

            # Add the Poisson innovation
            y_t_pred = carried_over + rng.poisson(mu_t)
            y_pred.append(y_t_pred)

            # Update the lagged counts for the next prediction
            last_p_y.pop(0)
            last_p_y.append(y_t_pred)

        return np.array(y_pred)  # Return predictions as a NumPy array
```

```python
# Helper function to calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # To avoid division by zero
    non_zero = y_true != 0
    return np.mean(np.abs((y_true[non_zero] - y_pred[non_zero]) / y_true[non_zero])) * 100
```

```python
# Helper function to calculate sMAPE
def symmetric_mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    non_zero = denominator != 0
    return np.mean(np.abs(y_true[non_zero] - y_pred[non_zero]) / denominator[non_zero]) * 100
```

```python
# Initialize a dictionary to store metrics
metrics_dict = {
    'Model': [],
    'MSE': [],
    'RMSE': [],
    'MAE': [],
    'MAPE': [],
    'sMAPE': []
}

# Calculate metrics for each model
models = ['Poisson', 'NegativeBinomial', 'Poisson_INAR']
for model in models:
    y_pred = predictions_df[model]
    y_true = predictions_df['Actual']

    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    smape = symmetric_mean_absolute_percentage_error(y_true, y_pred)

    metrics_dict['Model'].append(model)
    metrics_dict['MSE'].append(mse)
    metrics_dict['RMSE'].append(rmse)
    metrics_dict['MAE'].append(mae)
    metrics_dict['MAPE'].append(mape)
    metrics_dict['sMAPE'].append(smape)

# Create a DataFrame
metrics_df = pd.DataFrame(metrics_dict)
display(metrics_df)
```
